Hotel Chain Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data

df = pd.read_csv('state_count_results.csv')

# Set 'Place Name' column as index
df = df.set_index('Place Name')
```

Heatmap of hotel chains acrosss states

```python
plt.figure(figsize=(20,12))
sns.heatmap(df, cmap='YlOrRd', annot=False)
plt.title('Heatmap of the Hotel Chain Presence Across US States')
plt.tight_layout()
plt.savefig('heatmap_hotel_chains.png')
plt.show()
plt.close()
```

Bar Plot of total locations for each hotel chain

```python
total_locations = df.sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
total_locations.plot(kind='bar')
plt.title('Total Locations by Hotel Chains')
plt.xlabel('Hotel Chain')
plt.ylabel('Number of Locations')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('total_locations_bar.png')
plt.show()
plt.close()
```

Box plot of location distribution for each hotel chain

```python
plt.figure(figsize=(12,6))
df.T.boxplot()
plt.title('Distribution of Locations by Hotel Chain')
plt.xlabel('Hotel Chain')
plt.ylabel('Number of Locations')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('location_distribution_boxplot.png')
plt.show()
plt.close()
```

Stacked bar chart for top 5 hotel chains in top 10 states

```python
# Calculate total locations for each state and each hotel chain

chain_totals = df.sum(axis=1).sort_values(ascending=False)
state_totals = df.sum().sort_values(ascending=False)

# Select top 5 chains and top 10 states

top_5_chains = chain_totals.nlargest(5).index
top_10_states = state_totals.nlargest(10).index

# created a new dataframe with only top 5 chains and top 10 states

top_data = df.loc[top_5_chains, top_10_states]

plt.figure(figsize=(15, 8))
top_data.T.plot(kind='bar', stacked=True)
plt.title('Top 5 Hotel Chains in Top 10 States')
plt.xlabel('State')
plt.ylabel('Number of Locations')
plt.legend(title='Hotel Chain', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('top_5_chains_top_10_states.png')
plt.show()
plt.close()
```
